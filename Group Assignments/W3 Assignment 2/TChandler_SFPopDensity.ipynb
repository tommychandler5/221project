```python
df = pd.read_csv(
    'Data/SF_Pop_Density.csv',
    dtype=
    {
        'Geo_FIPS':str,
        'Geo_STATE':str,
        'Geo_COUNTY': str
    }
)
```

```python
df.head()
```

```python
df.info(verbose=True, show_counts=True)
```

```python
df.columns[df.isna().all()].tolist()
```

```python
df = df.dropna(axis=1,how="all")
```

```python
df.info()
```

```python
df.head()
```

```python
columns_to_keep = ['Geo_FIPS',
                   'SE_A00002_001',
                   'SE_A00002_002',
                   'SE_A00002_003']
```

```python
df2 = df[columns_to_keep]
```

```python
df2.head()
```

```python
df2.columns = ['FIPS',
              'TotalPop',
              'PopDensity',
              'Area']
```

```python
df2[df2.TotalPop <1]
```

```python
#Removing the census tracts in the Fallaron Islands, the California Coast Line, and SF Bay from the data set to focus on mainland
df3 = df2[df2.TotalPop >1]
df3
```

```python
df3.sample(5)
```

```python
df3['PopDensity'].mean()
```

```python
df3['Area'].mean()
```

```python
df_sorted = df3.sort_values(by='TotalPop',ascending = False)
```

```python
df_sorted.head()
```

```python
import geopandas as gpd
```

```python
tracts=gpd.read_file('Data/2020CensusTracts_SF.geojson')
```

```python
tracts.head()
```

```python
tracts.plot(figsize=(12,10))
```

```python
tracts.info(verbose=True, show_counts=True)
```

```python
tracts = tracts[['geoid', 'geometry']]
```

```python
tracts.head()
```

```python
tracts.columns = ['FIPS',
                  'geometry']
tracts.head()
```

```python
tracts_density=tracts.merge(df3,on="FIPS")
```

```python
tracts_density.head()
```

```python
tracts_density.plot(figsize=(12,10),
                    column='PopDensity',
                    legend=True,
                    scheme='NaturalBreaks')
```

```python
tracts_density.plot(figsize=(10,10),
                    column='PopDensity',
                    legend=True,
                    scheme='equal_interval')
```

```python
import folium
```

```python
m = folium.Map(location=[37.76,-122.45],
               zoom_start = 12,
               titles='CartoDB positron',
               attribution='CartoDB')
m
```

```python
# plotting choropleth over the base map
folium.Choropleth(
    geo_data=tracts_density,
    data=tracts_density,
    key_on='feature.properties.FIPS',
    columns=['FIPS', 'PopDensity'],
    fill_color='BuPu',
    line_weight=0.1,
    fill_opacity=0.8,
    line_opacity=0.2,
    legend_name='Population Density per Square Mile (2017-2021)').add_to(m)
m
```

```python
m.save('SF_PopulationDensity.html')
```
