### PLOTTING & MAPPING ###

Now I am going to upload the 2020 Census Tract geojson file for San Francisco. This will allow me to map the COVID-19 data.

```python
import folium
```

```python
# read the SF Census Tracts geojson file
tracts=gpd.read_file('week04 data/Census 2020_ Tracts for San Francisco (1).geojson')
```

```python
tracts.head()
```

```python
# plot it!
tracts.plot(figsize=(12,10))
```

```python
# I only need the FIPS and geometry information. In this set geoid is the same as the FIPS code.
tracts = tracts[['geoid','geometry']]
```

```python
# renaming the columns so that I can merge the COVID-19 data and census tract file using the FIPS column
tracts.columns = ['FIPS','geometry']
```

```python
# making sure that the columns were sucessfully renamed
tracts.head()
```

```python
#renaming the COVID-19 data file column from id to FIPS
df = df.rename(columns={'id': 'FIPS'})
```

```python
df.head()
```

```python
# create a new dataframe based on the join
tracts_COVID=tracts.merge(df,on="FIPS")
```

```python
# what does it look like now?
tracts_COVID.head()
```

```python
# plot it as a histogram with 50 bins showing the frequency of COVID-19 case counts.
df['deaths'].plot.hist(bins=50)
```

```python
# What is the mean number of COVID-19 related deaths?
df['deaths'].mean()
```

```python
# What is the median number of COVID-19 related deaths?
df['deaths'].median()
```

Now I will make a bar chart that shows the 5 census tracts in SF with the highest number of COVID-19 cases and highest number of COVID-19 deaths.

```python
df=df.sort_values(by='count',ascending = False)
```

```python
df.head(5).plot.barh(x='FIPS',
    y='count', 
    title='Top 10 Census Tracts in SF with Highest Number of COVID-19 Cases')
```

```python
df=df.sort_values(by='deaths',ascending = False)
```

```python
df.head(5).plot.barh(x='FIPS',
    y='deaths', 
    title='Top 10 Census Tracts in SF with Highest Number of COVID-19 Deaths')
```

```python
import folium
```

```python
# mpaping COVID case count by census tract
m = folium.Map(location=[37.7,-122.4],
zoom_start = 12,
tiles='CartoDB positron',
attribution='CartoDB')

# plot chorpleth over the base map
folium.Choropleth(
    geo_data=tracts_COVID, # geo data
    data=tracts_COVID, # data
    key_on='feature.properties.FIPS', # key, or merge column
    columns=['FIPS', 'count'], # [key, value]
    fill_color='BuPu',
    line_weight=0.1,
    fill_opacity=0.8,
    line_opacity=0.2, # line opacity (of the border)
    legend_name='COVID Count by Census Tract, SF').add_to(m)

# name on the legend color bar
m
```
